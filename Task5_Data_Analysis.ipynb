```python
import pandas as pd

# Load your CSV file (replace the filename with your actual CSV)
df = pd.read_csv('customers-1000.csv')

# Show the first few rows
df.head()
```

```python
# Check basic info about the dataset
df.info()

# See summary statistics (mean, min, max, etc.)
df.describe()

# Check number of rows and columns
print("Shape:", df.shape)

# Check for missing values
df.isnull().sum()

```

```python
sales_by_region = df.groupby('Region')['Sales'].sum()
print(sales_by_region)
```

```python
sales_by_region = df.groupby('Customer ID')['Sales'].sum()
print(sales_by_region)
```

```python
country_counts = df['Country'].value_counts()
print(country_counts)
```

```python
company_counts = df['Company'].value_counts().head(10)
print(company_counts)
```

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
year_counts = df['Year'].value_counts().sort_index()
print(year_counts)
```

```python
import matplotlib.pyplot as plt

# Bar chart: Number of records per country
country_counts.plot(kind='bar', title='Number of People by Country', xlabel='Country', ylabel='Count')
plt.show()

# Pie chart: Top 5 countries
country_counts.head(5).plot(kind='pie', autopct='%1.1f%%', title='Top 5 Countries')
plt.ylabel('')
plt.show()

# Line chart: Records per year
year_counts.plot(kind='line', marker='o', title='Records by Year', xlabel='Year', ylabel='Count')
plt.show()
```
